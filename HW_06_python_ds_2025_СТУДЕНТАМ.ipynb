## Задание 1 (12 баллов)
Каково соотношение ушедших и лояльных клиентов? Покажите это на графике и дайте комментарий по соотношению.

```python
#НАЧАЛО ВАШЕГО РЕШЕНИЯ
"""
Exited	Статус лояльности (1 — ушедший клиент, 0 — лояльный клиент)
"""
# агрегация всех клиентов по атрибуту Exited
loyal_churn_customers = df[["CustomerId", "Exited"]].groupby(by="Exited").agg(num_customers=("CustomerId", "count"))

# вычисление коэффициента "ушедшие-лояльные клиенты"
ratio_loyal_churn_customers = (loyal_churn_customers[loyal_churn_customers.index==1].num_customers.values / loyal_churn_customers[loyal_churn_customers.index==0].num_customers.values)[0].round(3) * 100

print(f"соотношение ушедших и лояльных клиентов находится на уровне {ratio_loyal_churn_customers}%")
#КОНЕЦ ВАШЕГО РЕШЕНИЯ
```

**Выводы:**


* Соотношение ушедших и лояльных клиентов показывает, что на каждого лояльного клиента приходится ~ 1 ушедший.
* Это значение может говорить о наличии проблем в услугах, предоставляемых клиентам Банка
